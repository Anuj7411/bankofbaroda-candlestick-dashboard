```python
# -----------------------
# Step 1: Import Libraries
# -----------------------

import pandas as pd           # For data handling
import plotly.graph_objects as go  # For interactive candlestick charts
import plotly.io as pio        # For saving HTML and rendering in notebook

print("✅ Libraries imported: pandas, plotly")
```

Step 1 — Import Libraries

In this step, we import all the Python libraries we’ll use throughout the project:

- **pandas**: to load, clean, and preprocess the CSV data.
- **plotly.graph_objects**: to create the candlestick chart.
- **plotly.io**: to save our interactive chart as an HTML file and render inside Jupyter.

Importing these at the very beginning ensures every subsequent cell has access to the required functionality.

```python
# -----------------------
# Step 2: Read the CSV file
# -----------------------

# Adjust the filename/path if needed
csv_path = "bankbaroda_stock_data.csv"

# Read the CSV into a pandas DataFrame
df_raw = pd.read_csv(csv_path)

# Show basic info about the data
print(f"✅ Loaded file: {csv_path}")
print("Shape:", df_raw.shape)
print("Columns:", df_raw.columns.tolist())

# Display first 5 rows as a visual check
df_raw.head()
```

## Step 2 — Read the CSV File

In this step we:

1. **Specified the path** to our Bank of Baroda CSV file (`bankbaroda_stock_data.csv`).
2. **Used `pd.read_csv()`** to load the file into a pandas DataFrame called `df_raw`.
3. **Printed basic information** (number of rows/columns, column names).
4. **Displayed the first 5 rows** of the dataset so we can visually inspect that the data loaded correctly.

This step confirms that we can successfully access the raw data before moving on to preprocessing.


```python
# -----------------------
# Step 3: Preprocess the Data
# -----------------------

df = df_raw.copy()  # work on a copy

# 1. Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=False)  # dayfirst=False matched your data

# 2. Set Date as index and sort
df = df.set_index('Date').sort_index()

# 3. Convert OHLC & Volume columns to numeric (strip commas just in case)
for col in ['Open', 'High', 'Low', 'Close', 'Volume']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col].astype(str).str.replace(',', ''), errors='coerce')

# 4. Drop any unnamed column (old index column)
for col in df.columns:
    if 'unnamed' in col.lower():
        df = df.drop(columns=[col])
        print(f"Dropped column: {col}")

# 5. Show summary info and first few rows as visual check
print("✅ Preprocessing complete.")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("\nIndex range:", df.index.min(), "to", df.index.max())

# Display first 5 rows of cleaned DataFrame
df.head()
```

## Step 3 — Preprocess the Data

In this step we clean and structure the raw dataset:

1. **Converted the `Date` column to datetime** with `pd.to_datetime()` (using `dayfirst=False` because it produced valid dates for all rows).
2. **Set the `Date` column as the index** of the DataFrame and sorted by date. This is essential for time-series plots like candlesticks.
3. **Converted the OHLC & Volume columns to numeric** types, stripping out any commas. This ensures the candlestick chart receives proper numeric data.
4. **Dropped any extra “Unnamed” column** that might be an artifact of CSV export.
5. **Printed a summary** and displayed the first few rows to visually confirm the cleaned dataset.

This gives us a ready-to-use DataFrame (`df`) for plotting the candlestick chart in the next step.


```python
# -----------------------
# Step 4: Create and Save the Candlestick Chart
# -----------------------

# 1. Build the candlestick trace
candlestick = go.Candlestick(
    x=df.index,
    open=df['Open'],
    high=df['High'],
    low=df['Low'],
    close=df['Close'],
    name='Bank of Baroda'
)

# 2. Create the figure and customize layout
fig = go.Figure(data=[candlestick])

fig.update_layout(
    title='Bank of Baroda Stock Prices – Candlestick Chart',
    xaxis_title='Date',
    yaxis_title='Price (INR)',
    xaxis_rangeslider_visible=False,  # hide the default range slider
    template='plotly_white',
    width=900,
    height=500
)

# 3. Show the interactive plot in the notebook
fig.show()

# 4. Save the chart as an interactive HTML file
output_html = "bankbaroda_candlestick_dashboard.html"
fig.write_html(output_html)
print(f"✅ Candlestick chart saved as: {output_html}")
```

## Step 4 — Create the Candlestick Chart

In this step we visualize the cleaned Bank of Baroda stock data:

1. **Created a Plotly `Candlestick` trace** using the `Open`, `High`, `Low`, and `Close` columns from our preprocessed DataFrame.
2. **Added the trace to a `Figure`** and customized the layout:
   - Title: *Bank of Baroda Stock Prices – Candlestick Chart*
   - X-axis: Date
   - Y-axis: Price (INR)
   - Removed the default range slider for a cleaner look.
3. **Displayed the interactive chart directly inside the notebook** with `fig.show()`.
4. **Saved the chart as an HTML file** (`bankbaroda_candlestick_dashboard.html`) so it can be opened in any browser or embedded in reports/presentations.

This gives you an interactive candlestick chart to explore the stock’s price movements visually.
